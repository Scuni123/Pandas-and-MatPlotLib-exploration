Now for the 4 way subplots. Need to look this up.

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(m12, bins=20)
axs[0, 0].set_title('2012 Militiary spending')
axs[0, 1].hist(m13, bins=20, color = 'green')
axs[0, 1].set_title('2013 Military spending')
axs[1, 0].hist(m14, bins=20, color = 'red')
axs[1, 0].set_title('2014 Military spending')
axs[1, 1].hist(m15, bins=20, color = 'orange')
axs[1, 1].set_title('2015 Military spending')

for ax in axs.flat:
    ax.set(xlabel='Spending as % of GDP', ylabel='Count')

for ax in axs.flat:
    ax.label_outer()
                
```
